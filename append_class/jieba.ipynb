```python
import jieba
import pandas as pd

out = []
data_label = []
csvfile = pd.read_csv('notice_append_v3.csv')

for i, token in enumerate(csvfile['context']):
    # print(token)
    print(' '.join(jieba.cut(token,cut_all=False)))
    out.append(' '.join(jieba.cut(token,cut_all=False)))
    data_label.append('NOT')
    
    

data = {}
data['label'] = data_label
data['context'] = out
data = pd.DataFrame(data)
data.to_csv("notice_append_v3_cut.csv", encoding = "utf-8-sig", index = False)
    
```
