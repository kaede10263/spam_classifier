```python
import pandas as pd
from tqdm import tqdm
from bs4 import BeautifulSoup
import requests
from lxml import etree

original_context = []
data_info = pd.read_csv('spam_data.csv')

data_info = data_info[:50000]

## 進度條
steps = len(data_info)
progress_bar = tqdm(range(steps))

for i in data_info['md5sum']:
    try:
        url = "http://125.227.221.217:5555/showOneEmail/%s"%(i)
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        dom = etree.HTML(str(soup))

        original_context.append((dom.xpath('//*[@id="p2"]')[0].text))
        progress_bar.update(1)
    except:
         original_context.append([])
    


# list_test = [i for i in range(len(data_info['md5sum']))]

data_info['original'] = original_context
data_info.head()
data_info.to_csv('generative_label/spam_data_orig_2.csv')
```
